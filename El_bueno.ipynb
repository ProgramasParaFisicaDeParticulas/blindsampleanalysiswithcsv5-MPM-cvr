```python
import numpy as np
import plotly.graph_objects as go
from scipy.special import wofz
from scipy.optimize import curve_fit

x = df_sd2["MT"].to_numpy()
nbins = 1000
range_x = (0, 120)

counts, bin_edges = np.histogram(x, bins=nbins, range=range_x)
bin_centers = 0.5*(bin_edges[:-1] + bin_edges[1:])
bin_width = bin_edges[1] - bin_edges[0]
sigma_y = np.sqrt(np.maximum(counts, 1))

def voigt(x, mu, sigma, gamma, A):
    """Voigt = Gauss(σ) ⊗ Lorentz(Γ), con factor de área A."""
    z = ((x - mu) + 1j*gamma) / (sigma*np.sqrt(2))
    return A * np.real(wofz(z)) / (sigma*np.sqrt(2*np.pi))

p0 = [bin_centers[counts.argmax()],
      np.std(x)/2,
      np.std(x)/2,
      counts.max()]

popt, pcov = curve_fit(
    voigt, bin_centers, counts,
    p0=p0, sigma=sigma_y, absolute_sigma=True, maxfev=10000
)
mu, sigma, gamma, A = popt
dmu, dsigma, dgamma, dA = np.sqrt(np.diag(pcov))


fwhm = 0.5346*gamma + np.sqrt(0.2166*gamma**2 + (2.355*sigma)**2)

print(f"Pico μ = {mu:.2f} ± {dmu:.2f} GeV")
print(f"Ancho natural Γ = {2*gamma:.2f} ± {dgamma:.2f} GeV")
#print(f"Resolución σ = {sigma:.2f} ± {dsigma:.2f} GeV")
print(f"FWHM_Voigt ≈ {fwhm:.2f} GeV")

x_fit = np.linspace(*range_x, 2_000)
y_fit = voigt(x_fit, *popt)

fig = go.Figure()

fig.add_bar(
    x=bin_centers, y=counts, width=bin_width,
    marker=dict(color="#1f77b4"), opacity=1, name="df_sd"
)

fig.add_scatter(
    x=x_fit, y=y_fit, mode="lines", line=dict(width=2),
    name="Ajuste Voigt"
)

fig.update_layout(
    bargap=0,
    title=(r"$\mathrm{Distribución\;MT}$"
           f"<br>Pico = {mu:.2f} GeV, FWHM ≈ {fwhm:.2f} GeV"),
    xaxis_title=r"$MT\;[\mathrm{GeV}]$",
    yaxis_title="Eventos/bin",
    width=700, height=700,
    template="plotly_white"
)
fig.show()
```
